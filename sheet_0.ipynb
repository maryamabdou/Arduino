The iris dataset contains three types of flowers which are setosa, versicolor and virginica. 

Each flower has four features:
sepal length, sepal width, petal length and petal width. 

Target refers to the type of flower, each of them has an index. 

0 --> setosa

1 --> versicolor 

2 --> virginica

```python
from sklearn import datasets
import numpy as np
import pandas as pd

iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
df.head()
```

**Sheet 0:**

1. Plot the histogram for each class

```python
from sklearn import datasets
import matplotlib.pyplot as plt

iris = datasets.load_iris()
colors= ['blue', 'red', 'green', 'yellow']

for i in range (3):
  plt.figure()
  for j in range (4):
    plt.hist(iris.data[iris.target==i, j], label=iris.feature_names[j], color=colors[j])
    plt.xlabel(iris.target_names[i]) 
    plt.legend(loc='upper right')
   
plt.show()
```

2. Use scatter plot to plot every 2 attributes together

```python
from sklearn import datasets
import matplotlib.pyplot as plt

iris = datasets.load_iris()
no = [0, 1 , 2 , 3]
for j in range (4):
  no.pop(0)
  for i in no:
    plt.figure()
    plt.scatter(iris.data[iris.target==0, j], iris.data[iris.target==0, i], color='blue', label=iris.target_names[0])
    plt.scatter(iris.data[iris.target==1, j], iris.data[iris.target==1, i], color='red', label=iris.target_names[1])
    plt.scatter(iris.data[iris.target==2, j], iris.data[iris.target==2, i], color='green', label=iris.target_names[2])
    plt.xlabel(iris.feature_names[j])
    plt.ylabel(iris.feature_names[i])
    plt.legend(loc='upper right') 
  
plt.show()
```

3. Use 3D scatter plot to plot every 3 attributes together

```python
from sklearn import datasets
import matplotlib.pyplot as plt

iris = datasets.load_iris()
no = [0, 1 , 2]
for j in range (2):
  no.pop(0)
  for i in no:
    fig = plt.figure(figsize = (8, 6))
    ax = plt.axes(projection ="3d")
    ax.scatter3D(iris.data[iris.target==0, j], iris.data[iris.target==0, i], iris.data[iris.target==0, i+1], color='blue', label=iris.target_names[0])
    ax.scatter3D(iris.data[iris.target==1, j], iris.data[iris.target==1, i], iris.data[iris.target==0, i+1], color='red', label=iris.target_names[1])
    ax.scatter3D(iris.data[iris.target==2, j], iris.data[iris.target==2, i], iris.data[iris.target==0, i+1], color='green', label=iris.target_names[2])
    ax.set_xlabel(iris.feature_names[j])
    ax.set_ylabel(iris.feature_names[i])
    ax.set_zlabel(iris.feature_names[i+1])
    plt.legend(loc='upper right')
  if j == 0:
    fig = plt.figure(figsize = (8, 6))
    ax = plt.axes(projection ="3d")
    ax.scatter3D(iris.data[iris.target==0, j], iris.data[iris.target==0, j+1], iris.data[iris.target==0, j+3], color='blue', label=iris.target_names[0])
    ax.scatter3D(iris.data[iris.target==1, j], iris.data[iris.target==1, j+1], iris.data[iris.target==0, j+3], color='red', label=iris.target_names[1])
    ax.scatter3D(iris.data[iris.target==2, j], iris.data[iris.target==2, j+1], iris.data[iris.target==0, j+3], color='green', label=iris.target_names[2])
    ax.set_xlabel(iris.feature_names[j])
    ax.set_ylabel(iris.feature_names[j+1])
    ax.set_zlabel(iris.feature_names[j+3])
    plt.legend(loc='upper right')

plt.show()
```

4. Show the correlation matrix between the attributes and comment on it

```python
from sklearn import datasets
import numpy as np
import pandas as pd

iris = datasets.load_iris()
iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris.corr()
```

**Sepal length:** has negative correlation between it and sepal width, also it has positive correlation between petal length and petal width.

**Petal length:** has negative correlation between it and sepal width, also it has positive correlation between sepaal length and petal width.

5. Use boxplots to show the quartiles and the median of the sepal length,
sepal width, petal length and petal width for each kind of flowers

```python
from sklearn import datasets
import matplotlib.pyplot as plt

iris = datasets.load_iris()
for i in range (3):
  plt.figure(figsize = (6, 4))
  plt.boxplot([iris.data[iris.target==i, 0], iris.data[iris.target==i, 1], iris.data[iris.target==i, 2], iris.data[iris.target==i, 3]])
  plt.xlabel(iris.target_names[i]) 

plt.show()
```
